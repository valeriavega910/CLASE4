3. Desarrollo del Programa:
Implementar el programa en Python para gestionar tareas utilizando solo listas y diccionarios.
Agregar al menos tres tareas distintas al programa utilizando el método insert() de la lista.
Cada tarea debe ser representada como un diccionario con las claves nombre, descripcion y estado.

```python
lista_de_tareas=[]
tarea1={"nombre":"obtencion de datos","descripcion":"extraer datos de diferentes fuentes","estado":"completo"}
tarea2={"control de calidad de los datos":"verificar la integridad de los datos extraidos","estado":"completo"}
tarea3={"nombre":"limpieza de datos","descripcion":"eliminar datos duplicados y inconsistentes","estado":"pendiente"}

lista_de_tareas.insert(0,tarea1)
lista_de_tareas.insert(1,tarea2)
lista_de_tareas.insert(2,tarea3)

print(lista_de_tareas)
```

Utilizar el método pop() de la lista para eliminar una de las tareas ingresadas.

```python
print(lista_de_tareas.pop(0))
```

Imprimir en pantalla las tareas restantes almacenadas en la lista, accediendo directamente a los elementos mediante sus índices.

```python
tareas_finales=lista_de_tareas[0:]
print(tareas_finales)
```
